### Loading the Image

```python
# %pip install pytesseract
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"D:\Software\tesseract.exe"
```

```python
from PIL import Image
from io import BytesIO

image = Image.open("../IMG_20240817_172625.png")
buffered = BytesIO()
image.save(buffered, format="PNG")
image_bytes = buffered.getvalue()
```

```python
print(image_bytes)
```

```python
from langchain_core.prompts import PromptTemplate

template = "Describe the image in two lines: {image}"
prompt = PromptTemplate(template = template, input_variables=["image"])

chain = prompt | llm
```

```python
result = chain.invoke({"image": image_bytes})
```
